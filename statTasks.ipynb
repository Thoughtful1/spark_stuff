```python
import pandas as pd
import numpy as np
```

```python
datasample = pd.read_csv('sample10000.csv', parse_dates=["tpep_pickup_datetime", "tpep_dropoff_datetime"])
```

```python
is_paid_with_cash = (datasample.payment_type == 2)


ph = is_paid_with_cash.mean()
s = np.sqrt(ph * (1-ph) / len(is_paid_with_cash))
s
```

```python
from statsmodels.stats.proportion import proportion_confint
proportion_confint(sum(is_paid_with_cash), len(is_paid_with_cash), alpha=0.01)
```

```python
avgdist =datasample.trip_distance.mean()
container11=datasample.trip_distance

standartdeviation = container11.std(ddof=1) / np.sqrt(len(container11))
from statsmodels.stats.weightstats import _tconfint_generic
_tconfint_generic(avgdist, 
                  standartdeviation, 
                  len(container11) - 1, 
                  0.05, 'two-sided')
```
